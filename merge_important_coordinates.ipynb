```python
import pandas as pd
```

```python
important_words = pd.read_parquet('Top10Words_DataFrame.parquet')
streetmapping = pd.read_excel('streetmapping.xlsx', index_col=0)
```

```python
important_words
```

```python
streetmapping
```

```python
# merge the two dataframes on location
merge = important_words.merge(streetmapping, left_on='Ort', right_on='location')
merge = merge.drop(columns=['location'])

# Convert "Erstellt", "Aktualisiert" and "Gelöst" to datetime with start date 1899-12-30 with accuracy of seconds
merge['Erstellt'] = pd.to_datetime(merge['Erstellt'], unit="D", origin=pd.Timestamp('1899-12-30')).dt.strftime('%Y-%m-%d %H:%M:%S')
merge['Aktualisiert'] = pd.to_datetime(merge['Aktualisiert'], unit="D", origin=pd.Timestamp('1899-12-30')).dt.strftime('%Y-%m-%d %H:%M:%S')
merge['Gelöst'] = pd.to_datetime(merge['Gelöst'], unit="D", origin=pd.Timestamp('1899-12-30')).dt.strftime('%Y-%m-%d %H:%M:%S')
merge
```

```python
merge.to_parquet('Top10Words_DataFrame_with_coordinates.parquet')
merge.to_excel('Top10Words_DataFrame_with_coordinates.xlsx')
```
